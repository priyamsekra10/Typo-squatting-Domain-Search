```python
import requests
import time
import string
from difflib import SequenceMatcher
import itertools
company_sectors = {
    "technology": ["tech", "io", "digital", "lab", "techies", "code", "software", "cloud", "help", "contact", "technologie", "numerique", "laboratoire", "logiciel", "informatique","-tech", "-io", "-digital", "-lab", "-techies", "-code", "-software", "-cloud", "-help", "-contact", "-technologie", "-numerique", "-laboratoire", "-logiciel", "-informatique"],
    "healthcare": ["med", "care", "health", "clinic", "wellness", "doctor", "medical", "pharma", "vital", "contact", "help", "sante", "clinique", "medecin", "medicament", "pharmacie", "bien-etre", "-med", "-care", "-health", "-clinic", "-wellness", "-doctor", "-medical", "-pharma", "-vital", "-contact", "-help", "-sante", "-clinique", "-medecin", "-medicament", "-pharmacie", "-bien-etre"],
    "finance": ["bank", "invest", "capital", "financial", "money", "wealth", "advisors", "fund", "crypto", "support", "help", "contact", "banque", "investissement", "finance", "argent", "conseiller", "fonds", "-bank", "-invest", "-capital", "-financial", "-money", "-wealth", "-advisors", "-fund", "-crypto", "-support", "-help", "-contact", "-banque", "-investissement", "-finance", "-argent", "-conseiller", "-fonds"],
    "retail": ["store", "shop", "market", "online", "buy", "sale", "deals", "e-commerce", "discount", "help", "contact", "magasin", "achat", "vente", "commerce", "rabais", "discount", "-store", "-shop", "-market", "-online", "-buy", "-sale", "-deals", "-e-commerce", "-discount", "-help", "-contact", "-magasin", "-achat", "-vente", "-commerce", "-rabais", "-discount"],
    "manufacturing": ["mfg", "industries", "factory", "production", "made", "build", "create", "manufacture", "supply", "contact", "help", "fabrication", "usine", "approvisionnement", "-mfg", "-industries", "-factory", "-production", "-made", "-build", "-create", "-manufacture", "-supply", "-contact", "-help", "-fabrication", "-usine", "-approvisionnement"],
    "energy": ["power", "energy", "solar", "renewable", "eco", "green", "electric", "wind", "efficiency", "help", "contact", "energie", "solaire", "renouvelable", "ecologique", "electrique", "eolien", "-power", "-energy", "-solar", "-renewable", "-eco", "-green", "-electric", "-wind", "-efficiency", "-help", "-contact", "-energie", "-solaire", "-renouvelable", "-ecologique", "-electrique", "-eolien"],
    "transportation": ["transport", "logistics", "ship", "express", "delivery", "move", "cargo", "fleet", "transporter", "contact", "help", "transport", "logistique", "livraison", "mouvement", "flotte", "-transport", "-logistics", "-ship", "-express", "-delivery", "-move", "-cargo", "-fleet", "-transporter", "-contact", "-help", "-transport", "-logistique", "-livraison", "-mouvement", "-flotte"],
    "hospitality": ["hotel", "travel", "event", "tour", "stay", "experience", "vacation", "destination", "resort", "help", "contact", "hotel", "voyage", "evenement", "sejour", "experience", "vacances", "-hotel", "-travel", "-event", "-tour", "-stay", "-experience", "-vacation", "-destination", "-resort", "-help", "-contact", "-hotel", "-voyage", "-evenement", "-sejour", "-experience", "-vacances"],
    "professional_services": ["consult", "law", "agency", "advisory", "expert", "service", "consulting", "solution", "partner", "contact", "help", "conseil", "juridique", "agence", "-consult", "-law", "-agency", "-advisory", "-expert", "-service", "-consulting", "-solution", "-partner", "-contact", "-help", "-conseil", "-juridique", "-agence"],
    "education": ["edu", "school", "learn", "academy", "knowledge", "study", "teach", "training", "tutor", "help", "contact", "education", "ecole", "apprendre", "formation", "enseigner", "tutoriel", "-edu", "-school", "-learn", "-academy", "-knowledge", "-study", "-teach", "-training", "-tutor", "-help", "-contact", "-education", "-ecole", "-apprendre", "-formation", "-enseigner", "-tutoriel"]
}


word = input("Enter the domain: ")
sector = input("Enter the sector: ")

# Splitting the word into prefix and domain
parts = word.split(".")
prefix = parts[0]
domain = "." + parts[1]

# Checking if the specified sector exists in the company_sectors dictionary
if sector in company_sectors:
    # Adding the corresponding keywords to the prefix based on the specified sector
    keywords = company_sectors[sector]
    modified_words = [prefix  + keyword + domain for keyword in keywords]
else:
    modified_words = []

# Adding the original word to the list
modified_words.append(word)

# print("Modified Words:")
# print(modified_words)
import string
from difflib import get_close_matches

def levenshtein_distance(s1, s2):
    if len(s1) > len(s2):
        s1, s2 = s2, s1

    distances = range(len(s1) + 1)
    for i2, c2 in enumerate(s2):
        distances_ = [i2 + 1]
        for i1, c1 in enumerate(s1):
            if c1 == c2:
                distances_.append(distances[i1])
            else:
                distances_.append(1 + min((distances[i1], distances[i1 + 1], distances_[-1])))
        distances = distances_
    return distances[-1]

def generate_typosquatting_domains(domain, sector):
    visually_similar_chars = {
        'o': ['0'],  # visually similar to zero
        '0': ['o'],  # visually similar to letter 'o'
        '1': ['l'],  # visually similar to letter 'l'
        'l': ['1'],  # visually similar to number 1
        'i': ['1', 'l'],  # visually similar to number 1 and letter 'l'
        'b': ['6', '8'],  # visually similar to number 6 and 8
        '6': ['b'],  # visually similar to letter 'b'
        '8': ['b'],  # visually similar to letter 'b'
        's': ['5'],  # visually similar to number 5
        '5': ['s'],  # visually similar to letter 's'
        'z': ['2'],  # visually similar to number 2
        '2': ['z'],  # visually similar to letter 'z'
        'g': ['q'],  # visually similar to letter 'q'
        'q': ['g'],  # visually similar to letter 'g'
        'n': ['m'],  # visually similar to letter 'm'
        'm': ['n'],  # visually similar to letter 'n'
        'v': ['u'],  # visually similar to letter 'u'
        'u': ['v'],  # visually similar to letter 'v'
        'd': ['cl'],  # visually similar to letters 'c' and 'l'
        'c': ['d'],  # visually similar to letter 'd'
        'l': ['cl'],  # visually similar to letters 'c' and 'l'
        't': ['f'],  # visually similar to letter 'f'
        'f': ['t'],  # visually similar to letter 't'
        'j': ['t'],  # visually similar to letter 't'
        'k': ['x'],  # visually similar to letter 'x'
        'x': ['k']
    }

    alphabet = string.ascii_lowercase + string.digits
    domains = set()
    removed_extension = ""

    # Remove any existing extension from the domain
    if '.' in domain:
        domain, extension = domain.rsplit('.', 1)
        removed_extension = "." + extension
    else:
        extension = ""

    for i in range(len(domain)):
        char = domain[i]

        if char in visually_similar_chars:
            similar_chars = visually_similar_chars[char]
            for repl_char in similar_chars:
                typo_domain = domain[:i] + repl_char + domain[i + 1:]
                domains.add(typo_domain)

        # Replace the character with visually similar characters based on Levenshtein distance
        matches = get_close_matches(char, alphabet, n=3, cutoff=0.5)
        for repl_char in matches:
            if repl_char != char:
                typo_domain = domain[:i] + repl_char + domain[i + 1:]
                domains.add(typo_domain)

    return domains, removed_extension

# Example usage
domain = word
# sector = input("Enter the sector: ")

company_sectors = {
    "technology": [".tech", ".io", ".com", ".net",".fr"],
    "healthcare": [".med", ".care", ".com", ".net",".fr"],
    "finance": [".bank", ".com", ".net",".fr"],
    "retail": [".store", ".shop", ".com", ".net",".fr"],
    "manufacturing": [".mfg", ".com", ".net",".fr"],
    "energy": [".power", ".energy", ".solar", ".com", ".net",".fr"],
    "transportation": [".transport", ".ship", ".com", ".net",".fr"],
    "hospitality": [".hotel", ".travel", ".com", ".net",".fr"],
    "professional_services": [".consult", ".law", ".com", ".net",".fr"],
    "education": [".edu", ".com", ".net",".fr"],
}

typosquatting_domains, removed_extension = generate_typosquatting_domains(domain, sector)

modified_words1 = []

for extension in company_sectors.get(sector, []):
    if extension != removed_extension:
        if domain.endswith(".fr"):
            modified_words1.append(domain[:-3] + extension)
        for typo_domain in typosquatting_domains:
            if domain.endswith(".fr"):
                modified_words1.append(typo_domain + extension)
            else:
                modified_words1.append(typo_domain + extension)

modified_words.append(domain)
# print("Modified Words:")
# print(modified_words1)
final = modified_words1 + modified_words
print(final)
len(final)
# final = final[:10]
```
